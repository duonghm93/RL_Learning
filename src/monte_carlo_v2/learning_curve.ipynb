```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('log.txt', sep=':', header=None)
df.columns = ['n_ep', 'num_step']
df['ep'] = df.n_ep.apply(lambda x: int(x.split(' ')[-1]))
df['ep_grp'] = df.ep.apply(lambda x: int(x/100))
```

```python
df_grp = df.groupby('ep_grp').agg({'num_step': 'mean'}).reset_index()
```

```python
fig, ax = plt.subplots(figsize=(20,5))
ax.plot(df.epp, df_grp.num_step, '--o')
plt.show()
plt.close()
```

```python
fig, ax = plt.subplots(figsize=(20,5))
ax.plot(df_grp.ep_grp, df_grp.num_step, '--o')
plt.show()
plt.close()
```
